```python
#Read in the data on hidden layer tuning which was recorded in parts
data1 = pd.read_table('../hw2_data.nosync/chart_data/gru_hidden.txt')
data2 = pd.read_table('../hw2_data.nosync/chart_data/gru_hidden_300.txt')
data3 = pd.read_table('../hw2_data.nosync/chart_data/gru_hidden500.txt')
```

```python
data = data1.columns.tolist() + data2.columns.tolist() + data3.columns.tolist()
```

```python
gru_hidden = {}

for i in data:
    gru_hidden.update(ast.literal_eval(i))
```

```python
def string_to_dict(data):
    '''Convert string to dictionary'''
    out = {}

    for i in data:
        out.update(ast.literal_eval(i))
    return out
```

```python
def make_plots(train_loss_hist, train_acc_hist,  val_loss_hist, val_acc_hist, model_type, iteration=300):
    '''Create plots of accuracy and loss'''
    #Accuracy
    plt.plot(train_acc_hist, label='Train Set Accuracy')
    plt.plot(val_acc_hist, label='Val Set Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Iteration % ' + str(iteration))
    plt.title(model_type + ' Training Curves')
    plt.legend()
    plt.show()
    
    #Loss
    plt.plot(train_loss_hist, label = 'Train Loss')
    plt.plot(val_loss_hist, label = 'Val Loss')
    plt.xlabel('Iteration % ' + str(iteration)) 
    plt.ylabel('Loss')
    plt.title(model_type + ' Training Loss per Iteration')
    plt.legend()
    plt.show()
    
    
    print( "Max Val Acc {}".format(max(val_acc_hist)))
    print('Max Val Acc Iteration {}'.format(np.argmax(val_acc_hist)*300))
    print( "Avg Val Acc {}".format(np.average(val_acc_hist)))
```

```python
def plot_tuning(parameters, val_acc, title):
    '''Single plot of tuning results'''
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_subplot(111)
    ax.plot(parameters, val_acc, 'o--')
    plt.title(title)
    plt.ylabel('Max Validation Accuracy')
    plt.show()
    
```

```python
def make_plots_param(d, model='GRU', param= 'Hidden Layers', iteration=300):
    max_val = []

    for n in d.keys():
        make_plots(d[n][0], d[n][1], d[n][2], d[n][3], model + ' ' + str(n) + ' ' + param, iteration=iteration)
        max_val.append(max(d[n][3]))
        
    return max_val
```

```python
max_val = make_plots_param(gru_hidden)
```

```python
plot_tuning(list(gru_hidden.keys()), max_val, 'GRU Hidden Layer Size Tuning')
```
